```python
import time
import sys
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://www.gametop.com/category/top.html')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
links=soup.find_all('div',attrs={'class':'row align-items-center bg-light rounded-lg box-shadow p-1 mb-2'})
```

```python
data_array=[]
for i in range(0,len(links)):
    print(links[i].text)
    data_array.append((links[i].text.replace('\n',' ')))
```

```python
data_array
```

```python
games_data=[]
for i in range(0,len(data_array)):
    array_data=(str(data_array[i]).split())
    Serial_No=int(array_data[0])
    Game=array_data[1]+" "+array_data[2]
    Rating=float(array_data[-3])
    Downloads=(array_data[-2])
    data=[Serial_No,Game,Rating,Downloads]
    games_data.append(data)
```

```python
games_data
```

```python
final_data_games=pd.DataFrame(games_data,columns=['SerialNo','GamesNames','Ratings','Downloads'])
```
